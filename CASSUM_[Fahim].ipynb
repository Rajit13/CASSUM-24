The carbon-to-oxygen (C/O) ratio and metallicity of a planetary atmosphere play crucial roles in understanding its formation history and composition.

1. **C/O Ratio**:
   - The C/O ratio indicates the abundance of carbon relative to oxygen in a planetary atmosphere.
   - It has significant implications for the types of molecules that can form in the atmosphere, as well as for the overall chemistry and climate of the planet.
   - A high C/O ratio favors the formation of carbon-rich molecules like carbon monoxide (CO) and methane (CH4), while a low C/O ratio favors oxygen-rich molecules like carbon dioxide (CO2) and water (H2O).

2. **Metallicity**:
   - Metallicity refers to the abundance of elements heavier than hydrogen and helium, collectively known as "metals," in a planetary atmosphere.
   - High metallicity suggests a greater abundance of elements like iron, silicon, and oxygen, which are crucial building blocks for rocky planets.
   - Metallicity can influence the formation and evolution of planetary systems, affecting processes like planet formation, atmospheric composition, and habitability.

**Relation to Formation History**:
   
- **Formation Environment**: The C/O ratio and metallicity of a planetary atmosphere are influenced by the composition of the protoplanetary disk from which the planet formed.
- **Protoplanetary Disk Composition**: The initial composition of the disk, including the relative abundances of volatile elements like carbon, oxygen, and metals, sets the stage for the chemical composition of the forming planet's atmosphere.
- **Migration and Accretion**: During planetary formation, migration and accretion processes can also affect the final C/O ratio and metallicity of a planet's atmosphere.
- **Stellar Influence**: The type and metallicity of the host star can also impact the composition of planetary atmospheres. Stars with higher metallicity are more likely to host planets with higher metallicity due to the availability of heavy elements in the protoplanetary disk.

In summary, the C/O ratio and metallicity of a planetary atmosphere provide valuable insights into its formation history, reflecting the interplay of processes such as disk chemistry, planet formation dynamics, and stellar characteristics.

According to ARCiS paper: In our definition of the metallicity this is the added number density of all elements heavier than He divided by the added number density of H and He.  Metallicity parameter, $[Z]$

$$[Z]=\log_{10}\left[\left(\frac{\sum_{i \neq H,He} X_i}{\sum_{i=H,He}X_i}\right)\cdot \left(\frac{\sum_{i = H,He} X_i}{\sum_{i\neq H,He}X_i}\right)_\odot \right]$$

with $X_i$ the number density of element $i$.

**Inside-Out Planet Formation**: radial drift of cm-m sized solids by gas drag is fast and leads to build up of a massive “pebble” ring at a location of a local pressure maximum. This location is likely to be the inner boundary of the dead zone. Planet formation occurs in this ring, perhaps via gravitational instability. The planet mass may grow until gap-opening, which then causes retreat of the dead zone inner boundary. A new pebble ring forms, slightly further out, and the process repeats. Or the planet Type I migrates inwards, allowing a new ring and planet to form behind it at the fixed dead zone inner boundary.

Features of this model:
- The radial drift of pebbles (part of the “meter-sized barrier” for planet formation) helps direct solids to the inner region for sequential planet formation.
- This process can create ~1-1$0M_\oplus$ planets on tightly-packed orbits close to the host star, starting from typical protoplanetary disk properties.
- It predicts relatively flat scalings of planetary mass with orbital radius, consistent with observed systems (for the scenario that assumes subsequent migration is negligible).
- Planetary orbital spacings should be >~3 Hill radii of the inner planet of the pair, and the spacing from first to second planet may be relatively larger than subsequent spacings, as is observed

## Update Solar Elementfile Values

We're using Metallicity as
$$[Z]=\log_{10}\left[\left(\frac{C+O}{H}\right)\cdot \left(\frac{C+O}{H}\right)_\odot^{-1} \right]$$

```python
import math

# Step 1: Store the initial data in a dictionary
element_values = {
    "H": 1.000E+00,
    "He": 7.722E-02,
    "C": 3.357E-04,
    "N": 9.594E-05,
    "O": 7.559E-04,
    "F": 3.537E-08,
    "Na": 2.338E-06,
    "Mg": 3.870E-05,
    "Al": 2.989E-06,
    "Si": 4.306E-05,
    "P": 3.037E-07,
    "S": 1.676E-05,
    "Cl": 3.032E-07,
    "K": 1.375E-07,
    "Ca": 2.347E-06,
    "Ti": 1.123E-07,
    "Cr": 5.169E-07,
    "Mn": 2.446E-07,
    "Fe": 2.406E-05
}

# Step 2: Define a function to modify the value
def modify_element_value(element, multiple):
    if element in element_values:
        element_values[element] *= multiple
    else:
        print(f"Element {element} not found in the dictionary.")

# Step 3: Function to calculate the metallicity
def calculate_metallicity():
    C_O_H = (element_values["C"] + element_values["O"]) / element_values["H"]
    C_O_H_sun = (3.357E-04 + 7.559E-04) / 1.000E+00
    metallicity = math.log10(C_O_H / C_O_H_sun)
    return metallicity

# Step 4: Print the initial metallicity
initial_metallicity = calculate_metallicity()
print(f"Initial Metallicity: {initial_metallicity:.3E}")

# Step 5: Modify the value of the specified element (e.g., Oxygen by 50 times)
element_to_modify = "O"
multiple = 50
modify_element_value(element_to_modify, multiple)

# Step 6: Print and save the updated values to a file
filename = f"solar_{element_to_modify}{multiple}.dat"
with open(filename, 'w') as file:
    for element, value in element_values.items():
        file.write(f"{element}: {value:.3E}\n")
        print(f"{element} {value:.3E}")

print(f"Updated values saved to {filename}")

# Step 7: Calculate and print the modified metallicity
modified_metallicity = calculate_metallicity()
print(f"Modified Metallicity: {modified_metallicity:.3E}")
```

```python
import math

# Step 1: Store the initial data in a dictionary
element_values = {
    "H": 1.000E+00,
    "He": 7.722E-02,
    "C": 3.357E-04,
    "N": 9.594E-05,
    "O": 7.559E-04,
    "F": 3.537E-08,
    "Na": 2.338E-06,
    "Mg": 3.870E-05,
    "Al": 2.989E-06,
    "Si": 4.306E-05,
    "P": 3.037E-07,
    "S": 1.676E-05,
    "Cl": 3.032E-07,
    "K": 1.375E-07,
    "Ca": 2.347E-06,
    "Ti": 1.123E-07,
    "Cr": 5.169E-07,
    "Mn": 2.446E-07,
    "Fe": 2.406E-05
}

# Step 2: Define a function to modify multiple element values
def modify_elements(modifications):
    for element, multiple in modifications.items():
        if element in element_values:
            element_values[element] *= multiple
        else:
            print(f"Element {element} not found in the dictionary.")

# Step 3: Function to calculate the metallicity
def calculate_metallicity():
    C_O_H = (element_values["C"] + element_values["O"]) / element_values["H"]
    C_O_H_sun = (3.357E-04 + 7.559E-04) / 1.000E+00
    metallicity = math.log10(C_O_H / C_O_H_sun)
    return metallicity

# Step 4: Print the initial metallicity
initial_metallicity = calculate_metallicity()
print(f"Initial Metallicity: {initial_metallicity:.3E}")

# Step 5: Define modifications for multiple elements (e.g., Oxygen by 50 times and Carbon by 1/50 times)
modifications = {
    "O": 20,
    "C": 1
}

# Step 6: Modify the elements
modify_elements(modifications)

# Step 7: Generate the filename based on the modifications
def format_multiple(multiple):
    if multiple >= 1:
        return f"{int(multiple)}"
    else:
        return f"d{int(1/multiple)}"

filename = "solar_" + "".join(f"{element}{format_multiple(multiple)}" for element, multiple in modifications.items()) + ".dat"

# Step 8: Print and save the updated values to a file
with open(filename, 'w') as file:
    for element, value in element_values.items():
        file.write(f"{element} {value:.3E}\n")
        print(f"{element} {value:.3E}")

print(f"Updated values saved to {filename}")

# Step 9: Calculate and print the modified metallicity
modified_metallicity = calculate_metallicity()
print(f"Modified Metallicity: {modified_metallicity:.3E}")
```

```python
import math

# Constants
G = 6.67430 * 10**-11  # Gravitational constant in m^3 kg^-1 s^-2
Jupiter_mass_kg = 1.898 * 10**27  # Mass of Jupiter in kg
Jupiter_radius_km = 71492  # Radius of Jupiter in km

# Given data
P_max = 3.6230 * 10**4  # Maximum pressure in Pascals
Rp = 0.1091 * Jupiter_radius_km * 1000  # Radius of the planet in meters
Mp_Jupiter_masses = 0.00943673  # Mass of the planet in Jupiter masses

# Convert planet mass to kg
Mp_kg = Mp_Jupiter_masses * Jupiter_mass_kg

# Function to calculate planet mass given maximum pressure
def calculate_planet_mass(P_max, Rp, Mp_kg):
    # Calculate gravitational acceleration g at the planet's surface
    g = G * Mp_kg / Rp**2

    # Mass of the atmosphere (1% of the planet's mass)
    Matm = 0.01 * Mp_kg

    # Calculate pressure using the ideal gas law (assuming surface pressure)
    P_calculated = (Matm * g) / (4 * math.pi * Rp**2)

    return P_calculated, Matm

# Calculate pressure and atmosphere mass
P_calculated, Matm = calculate_planet_mass(P_max, Rp, Mp_kg)

# Print the calculated pressure and compare with P_max
print(f"Calculated pressure at the planet's surface: {P_calculated:.2e} Pa")
print(f"Given maximum pressure (P_max): {P_max:.2e} Pa")

# Calculate the percentage of the planet's mass that is atmosphere
percentage_atmosphere_mass = (Matm / Mp_kg) * 100

# Print the percentage of the planet's mass that is atmosphere
print(f"The percentage of the planet's mass that is atmosphere: {percentage_atmosphere_mass:.2f}%")
```
